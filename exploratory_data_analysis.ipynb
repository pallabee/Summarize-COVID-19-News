### Analysis of the text length of data and target variables

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
X = clean.X
Y = clean.Y
```

```python
df = pd.DataFrame()
df['data_text_len'] = [len(text.split()) for text in X]
df['target_text_len'] = [len(text.split()) for text in Y]
```

```python
df.head()
```

```python
fig,axes = plt.subplots(1,2, figsize=(15,5), sharey = True)
sns.histplot(ax=axes[0], x = df['data_text_len'],bins=100)
axes[0].axvline(df['data_text_len'].mean(), color = 'red', linestyle = '--',label = 'Mean')
axes[0].axvline(df['data_text_len'].max(), color = 'green', linestyle = '--',label = 'Max')
axes[0].set_title("Mean: {:.2f}, Max: {}".format(df['data_text_len'].mean(),df['data_text_len'].max()))
axes[0].legend()
sns.histplot(ax=axes[1], x = df['target_text_len'],bins=100)
axes[1].axvline(df['target_text_len'].mean(), color = 'red', linestyle = '--',label = 'Mean')
axes[1].axvline(df['target_text_len'].max(), color = 'green', linestyle = '--',label = 'Max')
axes[1].set_title("Mean: {:.2f}, Max: {}".format(df['target_text_len'].mean(),df['target_text_len'].max()))
axes[1].legend()
fig.suptitle("Histogram showing distribution of text length of data and target variables",fontsize = 15)
plt.show()
```

```python
fig, axes = plt.subplots(1,2, figsize=(10, 10))
sns.boxplot(y = "data_text_len",data=df, color='blue', ax=axes[0])
axes[0].set(xlabel="data_text_len", ylabel="Count")
sns.boxplot(y = "target_text_len",data=df, color='yellow', ax=axes[1])
axes[1].set(xlabel="target_text_len", ylabel="Count")
fig.suptitle("Boxplot showing distribution of text length of data and target variables")
plt.show()
```

Observations from the above plots:
- The data and target contains empty strings(i.e text length = 0), which needs to be removed.
- For preparation of training dataset, maximum text size limit may be set (data text size = 500, target text size = 20), as most of the observations appear to be lying in this range.
